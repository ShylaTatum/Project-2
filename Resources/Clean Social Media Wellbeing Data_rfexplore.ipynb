To determine which social media platforms have the most engagement, we can analyze several metrics that contribute to overall engagement: daily usage, posts per day, likes received, comments received, and messages sent. We will calculate the average of these engagement metrics for each platform and then rank the platforms based on their total engagement scores.



```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the updated dataset
data_path = '/Users/rygofoss/Desktop/AI_SocialMediaEmotion Project2/train_update.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset to ensure it is correctly loaded
print(df.head())

# Drop rows with missing values in relevant columns
df = df.dropna(subset=['Dominant_Emotion', 'Posts_Per_Day', 'Likes_Received_Per_Day', 
                       'Comments_Received_Per_Day', 'Messages_Sent_Per_Day'])

# Calculate average engagement metrics per emotion
emotion_engagement = df.groupby('Dominant_Emotion')[['Posts_Per_Day', 
                                                     'Likes_Received_Per_Day', 
                                                     'Comments_Received_Per_Day', 
                                                     'Messages_Sent_Per_Day']].mean()

# Combine all engagement metrics into a total engagement score
emotion_engagement['Total_Engagement'] = emotion_engagement.sum(axis=1)

# Display the combined engagement data by emotion
print("\nAverage Engagement Metrics by Dominant Emotion:")
print(emotion_engagement)

# Plotting a scatter plot to show emotions based on levels of engagement
plt.figure(figsize=(14, 8))
plt.scatter(emotion_engagement['Total_Engagement'], emotion_engagement.index, 
            s=emotion_engagement['Likes_Received_Per_Day']*50,  # Bubble size based on likes received
            alpha=0.7, c=emotion_engagement['Comments_Received_Per_Day'], cmap='viridis')

# Add labels and titles
plt.title('Emotions Based on Levels of Engagement')
plt.xlabel('Total Engagement Score')
plt.ylabel('Dominant Emotion')
plt.colorbar(label='Average Comments Per Day')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
